```python
def factorial(n):
    prod = 1
    while n >= 1:
        prod = prod * n
        n = n - 1
    return prod
```

Now, create a function for the Binomial distribution with three arguments $n$, $p$ and $k$ just like in the formula:

$$ \large P(Y=k)= \binom{n}{k} p^k(1-p)^{(n-k)}$$ 


```python
def factorial(n):
    prod = 1
    while n >= 1:
        prod = prod * n
        n = n - 1
    return prod

def binom_distr(n, p, k):
    binom_coeff = factorial(n) / (factorial(k) * factorial(n - k))
    probability = binom_coeff * (p ** k) * ((1 - p) ** (n - k))
    return probability

# Example usage
n = 10
p = 0.25
k = 3

# Calculate the probability of getting exactly 3 strikes in 10 rounds
p_3_strikes = binom_distr(n, p, k)
print(f"The probability of getting exactly 3 strikes in 10 rounds is {p_3_strikes:.7f}")
```

## Visualize these results

Create the PMF using these empirical results (that is, the proportions based on the values we obtained running the experiment 5000 times).

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Set the random seed for reproducibility
np.random.seed(123)

# Number of simulations
num_simulations = 5000

# Parameters for the binomial distribution
n = 10  # Number of trials
p = 0.25  # Probability of success

# Perform the simulations
results = np.random.binomial(n, p, num_simulations)

# Count the occurrences of each outcome
outcomes, counts = np.unique(results, return_counts=True)

# Calculate the proportions (PMF)
pmf = counts / num_simulations

# Plot the PMF
plt.figure(figsize=(10, 6))
plt.bar(outcomes, pmf, width=0.6, color='skyblue', alpha=0.7)
plt.xlabel('Number of Strikes')
plt.ylabel('Probability')
plt.title('Probability Mass Function (PMF) for Bowling Strikes')
plt.xticks(outcomes)
plt.grid(axis='y')
plt.show()
#
#
#
#
```

You should see that, with a 25% strike hit rate, even when simulating 5000 times, an almost perfect and/or perfect game of 9 and 10 strikes didn't even occur once! If you change the random seed, however, you'll see that perfect games will show up occasionally. 

Next, let's create the CDF based on these results. You can use `np.cumsum` to obtain cumulative probabilities.

```python
# Your code here
## Calculate the cumulative probabilities (CDF)
cdf = np.cumsum(pmf)

# Plot the CDF
plt.figure(figsize=(10, 6))
plt.plot(outcomes, cdf, marker='o', linestyle='-', color='skyblue')
plt.xlabel('Number of Strikes')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF) for Bowling Strikes')
plt.xticks(outcomes)
plt.grid(axis='y')
plt.show()

#
#
#
```
